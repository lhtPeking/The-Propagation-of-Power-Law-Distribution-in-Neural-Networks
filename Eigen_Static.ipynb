```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import ConnectomeAnalysis, PlotMethod
```

```python
## -- Read CSV -- ##
df_Retina = pd.read_csv('./Connectome_Data/Mouse_retina.csv', header=None)
# print(df_Retina.head())
print("Mouse Retina csv Shape: ", df_Retina.shape)

df_Celegans = pd.read_csv('./Connectome_Data/Celegans.csv', header=None)
# print(df_Celegans.head())
print("Celegans csv Shape: ", df_Celegans.shape)

df_Drosophila_central_brain = pd.read_csv('./Connectome_Data/Drosophila_central_brain.csv', header=None)
# print(df_Drosophila_central_brain.head())
print("Drosophila_central_brain csv Shape: ", df_Drosophila_central_brain.shape)

df_Drosophila_optic_medulla = pd.read_csv('./Connectome_Data/Drosophila_optic_medulla.csv', header=None)
# print(df_Drosophila_optic_medulla.head())
print("Drosophila_optic_medulla csv Shape: ", df_Drosophila_optic_medulla.shape)

df_Platynereis_sensory_motor = pd.read_csv('./Connectome_Data/Platynereis_sensory_motor.csv', header=None)
# print(df_Platynereis_sensory_motor.head())
print("Platynereis_sensory_motor csv Shape: ", df_Platynereis_sensory_motor.shape)
```

## drosophila_optic_medull

```python
## -- construct connection matrix -- ##
connection_matrix_drosophila_optic_medulla = ConnectomeAnalysis.construct_connection_matrix('./Connectome_Data/Drosophila_optic_medulla.csv', make_dense=True)
print("drosophila_optic_medull connection matrix shape:", connection_matrix_drosophila_optic_medulla.shape)

## -- compute eigenspectrum -- ##
eigenvalue_list_drosophila_optic_medulla = ConnectomeAnalysis.compute_eigenspectrum(connection_matrix_drosophila_optic_medulla, symmetrization=True)

## -- compute shuffled eigenspectrum -- ##
eigenvalue_list_shuffled_drosophila_optic_medulla = ConnectomeAnalysis.compute_shuffled_eigenspectrum(connection_matrix_drosophila_optic_medulla, symmetrization=True, shuffle_diagonal=True, seed=42)

## -- compute brunel weighted eigenspectrum -- ##
brunel_weighted_eigenvalue_list_drosophila_optic_medulla = ConnectomeAnalysis.compute_brunel_weight_eigenspectrum(connection_matrix_drosophila_optic_medulla, symmetrization=True, neg_fraction=1/5, scale=4.0, seed=42)
```

```python
## -- plot and regession -- ##
_ = PlotMethod.log_log_plot(
        connection_matrix_drosophila_optic_medulla, # list or matrix 
        bin_num=200, 
        regression=True,
        range_=None, # tuple: (min,max)
        lowerbound=0.1,
        higherbound=0.9,
        num_points=500, # interception point number
        density=True,
        color_scatter="black",
        color_fit="grey",
        show=True,
        ax=None,
        return_params=True,
        xlabel="Log(Connection_Strength)"
        )
```

```python
## -- plot and regession -- ##
slope1, intercept1, ax_1 = PlotMethod.log_log_plot(
        eigenvalue_list_drosophila_optic_medulla, # list or matrix 
        bin_num=1000, 
        regression=True,
        range_=None, # tuple: (min,max)
        lowerbound=0.1,
        higherbound=0.9,
        num_points=1000, # interception point number
        density=True,
        color_scatter="darkblue",
        color_fit="lightblue",
        show=False,
        ax=None,
        return_params=True,
        xlabel="Log(Eigenvalue)"
        )

_ = PlotMethod.log_log_plot(
        eigenvalue_list_shuffled_drosophila_optic_medulla, # list or matrix 
        bin_num=500, 
        regression=True,
        range_=None, # tuple: (min,max)
        lowerbound=0.1,
        higherbound=0.9,
        num_points=1000, # interception point number
        density=True,
        color_scatter="green",
        color_fit="green",
        show=False,
        ax=ax_1,
        return_params=True,
        xlabel="Log(Eigenvalue)"
        )



_ = PlotMethod.log_log_plot(
        brunel_weighted_eigenvalue_list_drosophila_optic_medulla, # list or matrix 
        bin_num=1000, 
        regression=True,
        range_=None, # tuple: (min,max)
        lowerbound=0.1,
        higherbound=0.9,
        num_points=1000, # interception point number
        density=True,
        color_scatter="red",
        color_fit="salmon",
        show=True,
        ax=ax_1,
        return_params=True,
        xlabel="Log(Eigenvalue)"
        )
```
